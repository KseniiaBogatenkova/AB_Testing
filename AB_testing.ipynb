# Отчет по A/B-тестированию в рамках игры


## Цель игры:
Оптимизация выручки компании с использованием A/B-тестирования. 

## Стратегия принятия решений:
Моя основная стратегия основывалась на анализе P-value, прироста конверсии и логическом понимании того, как изменение может повлиять на поведение пользователей. Я ориентировалась на следующие принципы:
- Принятие в релиз тех экспериментов, где P-value ≤ 0.05 и наблюдался значимый прирост конверсии.
- Остановка экспериментов с высоким P-value (≥ 0.1) или при отсутствии значимого прироста конверсии.
- Внимание к контексту изменения (я старалась разнообразить работу таким образом, чтобы экспериметы шли каждый день)

### Сложные ситуации и сомнения:
 Были эксперименты, где P-value находился в "серой зоне" (≥ 0.05, но < 0.1). Признаюсь, были сомнения, продолжать их до получения большего объема данных, или принимать решение сразу.  
 Иногда прирост в конверсии был минимальным (≤ 0.2%), и я задавалась вопросом, достаточно ли этого для релиза.

## Рекомендации для будущей работы
1. Более гибкий подход к "серой зоне": эксперименты с P-value около 0.05 можно продолжать, чтобы собрать больше данных, особенно если прирост в конверсии наблюдается.
2. Фокус на влиянии изменений: учитывать, как конкретный вид изменения (дизайн, поиск, рекомендации) может влиять на долгосрочное поведение пользователей.
3. Учет абсолютных показателей: не только относительное улучшение, но и общий объем прироста покупателей имеет значение для финальной оценки.
4. Планирование параллельных тестов: минимизация пересечений экспериментов, чтобы не размывать данные.
